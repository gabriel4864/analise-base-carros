```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# LÃª o arquivo CSV
df = pd.read_csv("car_prices - car_prices.csv").copy()
print(df.head(5))
```

```python
# Removendo as linhas com dados faltantes
df = df.dropna()
```

```python
# Usando lambda expressions para filtrar os dados de 2010 
df_2010 = df[df['year'].apply(lambda x: x == 2010)]
print(df_2010.head(5))
```

```python
# Contando as marcas mais vendidas de 2010
top_5_marcas = (
    df_2010['make']
    .value_counts()
    .head(5)
)

df_plot = top_5_marcas.reset_index()
df_plot.columns = ["Make", "Quantidade de Vendas"]

sns.set()
fig, ax = plt.subplots(figsize = (15,5))
cores = ["#562155", "#72577c","#8e8ca3", "#7fb8b0" , "#a9c2c9", ]
g = ax.bar(df_plot["Make"], df_plot["Quantidade de Vendas"], color = cores, width = .5)
ax.set_title(" Marcas de carro mais vendidas de 2010")
plt.xlabel("Marca")
plt.ylabel("Quantidade de Vendas")
ax.bar_label(g, padding= - 20, color = "white")
ax.set_ylim((0, 4000))
plt.show()
```

```python
# Top 5 modelos mais vendidos em 2010
top_5_models = (
    df_2010['model']
    .value_counts()
    .head(5)
)

df_model = top_5_models.reset_index()
df_model.columns = ["Model", "Quantidade de Vendas"]

sns.set()
fig, ax = plt.subplots(figsize = (15,5))
cores = ["#2e0331", "#8a034d","#368986", "#0ba18c", "#5adb94" ]
g = ax.bar(df_model["Model"], df_model["Quantidade de Vendas"], color = cores, width = .5)
ax.set_title(" Modelos mais vendidos de 2010")
plt.xlabel("Modelo")
plt.ylabel("Quantidade de Vendas")
ax.bar_label(g, padding= - 20, size = 14, color = "white")
ax.set_ylim((0, 1500))
plt.show()
```

```python
# Top 5 tipos mais vendidos em 2010
top_2_body = (
    df_2010['body']
    .value_counts()
    .head(2)
)

df_body = top_2_body.reset_index()
df_body.columns = ["Body", "Quantidade de Vendas"]

sns.set()
fig, ax = plt.subplots(figsize = (15,5))
cores = ["#e81e4a", "#0b1d21"]
g = ax.bar(df_body["Body"], df_body["Quantidade de Vendas"], color = cores, width = .5)
ax.set_title(" Tipos de carro mais vendidos de 2010")
plt.xlabel("Tipo")
plt.ylabel("Quantidade de Vendas")
ax.bar_label(g, padding= - 40, size = 20, color = "white")
ax.set_ylim((0, 10000))
plt.show()
```
