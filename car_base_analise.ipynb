```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_excel("car_base_tratada.xlsx")

# Contando as marcas mais vendidas
top_5_marcas = (
    df['Brand']
    .value_counts()
    .head(5)
)

df_plot = top_5_marcas.reset_index()
df_plot.columns = ["Brand", "Quantidade de Vendas"]

sns.set()
fig, ax = plt.subplots(figsize = (15,5))
cores = ["#562155", "#72577c","#8e8ca3", "#7fb8b0" , "#a9c2c9", ]
g = ax.bar(df_plot["Brand"], df_plot["Quantidade de Vendas"], color = cores, width = .5)
ax.set_title(" Marcas de carro mais vendidas")
plt.xlabel("Marcas")
plt.ylabel("Quantidade de Vendas")
ax.bar_label(g, padding= - 20, color = "white")
ax.set_ylim((0, 1100))
plt.show()
```

```python
# Contando os modelos mais vendidos
top_5_modelos = (
    df['Model']
    .value_counts()
    .head(5)
)

df_plot = top_5_modelos.reset_index()
df_plot.columns = ["Model", "Quantidade de Vendas"]

sns.set()
fig, ax = plt.subplots(figsize = (15,5))
cores = ["#562155", "#72577c","#8e8ca3", "#7fb8b0" , "#a9c2c9", ]
g = ax.bar(df_plot["Model"], df_plot["Quantidade de Vendas"], color = cores, width = .5)
ax.set_title(" Marcas de carro mais vendidas")
plt.xlabel("Modelos")
plt.ylabel("Quantidade de Vendas")
ax.bar_label(g, padding= - 20, color = "white")
ax.set_ylim((0, 400))
plt.show()
```

```python
modelo_mais_caro = df.loc[df["Price"].idxmax()]
print(modelo_mais_caro)
```

```python
modelo_mais_barato = df.loc[df["Price"].idxmin()]
print(modelo_mais_barato)
```

```python
# Transmisses mais usadas

# Construo do grfico
transmissoes_mais_usadas = df["Transmission"].value_counts()
cores = ["#562155", "#8e8ca3", "#469187"]
plt.figure(figsize=(6, 6))
sns.set()
plt.pie(
    transmissoes_mais_usadas.values,
    labels=transmissoes_mais_usadas.index,
    colors=cores,
    autopct='%1.1f%%',
    startangle=90,
    textprops={
        "color": "black",   
        "fontsize": 13,
    }
)
plt.title(
    "Transmisses mais usadas",
    fontsize=18,           
    fontweight="bold",       
)
plt.show()
```

```python
# Transmisses mais usadas

# Construo do grfico
Combustiveis_mais_usados = df["Fuel type"].value_counts()
cores = ["#562155", "#779097", "#8e8ca3", "#469187"]
plt.figure(figsize=(6, 6))
sns.set()
plt.pie(
    Combustiveis_mais_usados.values,
    labels=Combustiveis_mais_usados.index,
    colors=cores,
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops=dict(width=0.5),
    pctdistance=0.75, 
    textprops={
        "color": "black",   
        "fontsize": 13,
    }
)
plt.title(
    "Combustveis mais usados",
    fontsize=18,           
    fontweight="bold",       
)
plt.show()
```
